```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.info()
```

```python
cleaned_df = df.dropna()
```

```python
cleaned_df.info()
```

```python
#Our NaN data was as '-', here we are droping all rows that are with '-':
for column in cleaned_df:
    cleaned_df = cleaned_df[cleaned_df[column] != ' - ']
```

```python
cleaned_df
```

#### Creating a new column that classifies avg2020 to groups by PM2.5 values:
1 = Good (10.1-12.0),
2 = Moderate (12.1-35.4),
3 = Unhealthy For Sensitive Groups (35.5-55.3),
4 = Unhealthy (55.4-150.3),
5 = Very Unhealthy(150.4-250.3).

```python
bins = [0,10.1, 12.1, 35.5, 55.4, 150.4, 250.4]
labels = [1, 2, 3, 4, 5,6]
cleaned_df["Classify Groups"] = pd.cut(cleaned_df["Avg 2020"], bins, labels=labels)
```

```python
cleaned_df.to_csv('cleaned_data.csv', encoding='utf-8', index=False)
```

Now we are spliting the City and Country to 2 different columns and re-aranging the DataFrame,
we want the country name to be beside the city name:

```python
cleaned_df[['city', 'Country']] = cleaned_df['city'].str.split(', ', 1, expand=True)
cleaned_df
```
