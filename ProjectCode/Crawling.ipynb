```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
for i in range(70, 90):
    url = 'https://www.iqair.com/world-most-polluted-cities?page=' + str(i) + '&perPage=50&cities='
    html = requests.get(url)
    with open('./data/page' + str(i) + '.html','wb') as fd:
        fd.write(html.content)
```

```python
cities = []
avg_2020 = []
january = []
february = []
march = []
april = []
may = []
june = []
july = []
august = []
september = []
october = []
november = []
december = []
avg_2019 = []
for i in range(1, 90):
    f = open("./data/page" + str(i) + ".html", "rb")
    file = f.read()
    soup = BeautifulSoup(file,"html.parser")
    table = soup('table',attrs={"class":"mat-table cdk-table mat-elevation-z8"})[0]
    for row in table("tr"):
        cells = row("td")
        if(len(cells) < 16):
            continue
        cities.append(cells[1].get_text())
        avg_2020.append(cells[2].get_text())
        january.append(cells[3].get_text())
        february.append(cells[4].get_text())
        march.append(cells[5].get_text())
        april.append(cells[6].get_text())
        may.append(cells[7].get_text())
        june.append(cells[8].get_text())
        july.append(cells[9].get_text())
        august.append(cells[10].get_text())
        september.append(cells[11].get_text())
        october.append(cells[12].get_text())
        november.append(cells[13].get_text())        
        december.append(cells[14].get_text())        
        avg_2019.append(cells[15].get_text())
```

```python
df = pd.DataFrame({ "city":cities, "Avg 2020":avg_2020, "January":january,
              "February":february, "March":march, "April":april, "May": may, "June": june, "July": july,
                   "August": august, "September": september, "October": october, "November": november, "December": december,
                   "Avg 2019": avg_2019
})
df
```

```python
df.to_csv('data.csv', encoding='utf-8', index=False)
```
